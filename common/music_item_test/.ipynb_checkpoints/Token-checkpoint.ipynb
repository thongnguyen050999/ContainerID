```python
pretty = pretty_midi.PrettyMIDI(resolution=480,initial_tempo=120)
program = pretty_midi.instrument_name_to_program('Acoustic Grand Piano')
piano = pretty_midi.Instrument(program=program)
```

```python
from common.constants import CONVERSION
files=['[21 Guns][Green Day][verse].xml',"['74-'75][The Connells][chorus].xml"]
for file in files:
    
    xmlItem = XMLItem()
    xmlItem.from_file(os.path.join('XML',file))
    remiItem = RemiItem(xmlItem)
    
    pretty = pretty_midi.PrettyMIDI(resolution=480,initial_tempo=120)
    program = pretty_midi.instrument_name_to_program('Acoustic Grand Piano')
    piano = pretty_midi.Instrument(program=program)

    notes = remiItem.to_notesequence()
    for note in notes:
        start = note.start
        end = note.start + note.duration
        start = CONVERSION.duration_to_quarterLength(start)
        end = CONVERSION.duration_to_quarterLength(end)
        start *= 480
        end *= 480
        start = pretty.tick_to_time(int(start))
        end = pretty.tick_to_time(int(end))
        print(start,end)
        n = pretty_midi.Note(velocity=90,pitch=note.pitch,start=start,end=end)
        piano.notes.append(n)

    pretty.resolution=480
    pretty.time_signature_changes.append(pretty_midi.TimeSignature(4,4,time=0))
    pretty.key_signature_changes.append(pretty_midi.KeySignature(14,0))
    pretty.instruments.append(piano)
    pretty.write(file+'.midi')
```
