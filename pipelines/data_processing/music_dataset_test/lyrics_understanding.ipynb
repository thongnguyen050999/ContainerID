```python
#create a DataFrame containing ssm data
ssm_df = pd.DataFrame(ssm_list)
```

```python
#Check types of data
line_errors = []
segment_errors = []
id_type = []
line_type = None
segment_type = None
id_type = None
for i,row in ssm_df.iterrows():
    if line_type is None:
        line_type = type(row['line'])
    if segment_type is None:
        segment_type = type(row['segment'])
    if id_type is None:
        id_type = type(row['id'])
    if not isinstance(row['line'],line_type):
        print("Inconsistent type at row {}".format(i))
        line_errors.append(i)
    if not isinstance(row['segment'],segment_type):
        print("Inconsistent type at row {}".format(i))
        segment_errors.append(i)
    if not isinstance(row['id'],id_type):
        print("Inconsistent type at row {}".format(i))
        id_errors.append(i)
    
        
```

```python
counter = dict()
num_samples = 500
for index,row in wasabi.iterrows():
    try:
        lyrics = lyricwikia.get_lyrics(row['artist'],row['title'])
        if lyrics is None or not isinstance(lyrics,str) or lyrics == "":
            continue
    except:
        continue
    paragraphs = lyrics.split("\n\n")
    segment_lyrics = [x.split("\n") for x in paragraphs]
    score = sum([len(x) for x in segment_lyrics])
    if score not in counter:
        counter[score] = 0
    counter[score] = counter[score] + 1
    if num_samples == index:
        break
    print("OK")

    
```
