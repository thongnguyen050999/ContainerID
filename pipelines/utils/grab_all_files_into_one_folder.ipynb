```python
import fnmatch
import os
import shutil
from pathlib import Path

def get_all_files_in_dir_recursively(directory, ext_list=['midi','mid','mxl']):
    src = Path(directory)
    matches = []
    for ext in ext_list:
        matches += src.glob('**/*.' + ext)
                
    return matches

def copy_all_files_in_one_folder(file_paths, destination_dir, name_process=None):
    for index, file_path in enumerate(file_paths):
        copy_file(file_path, destination_dir, name_process)
        
def copy_file(src, dest, name_process):
    if not os.path.isdir(dest):
        os.makedirs(dest)
       
    dest_file_path = name_process(src, dest)
    try:
    shutil.copyfile(str(src), str(dest_file_path))
    except:
        print('Error copying file path: ' + str(src))
        
src = Path(r"D:\Projects\viMusic\dataset\Pop\XML")
dest = Path(r"D:\Projects\viMusic\dataset\Pop\pop_hkt")
```

```python
def custom_name_ext_process(file_path, dest_dir):
    ##### Init
    src = Path(file_path)
    dest = Path(dest_dir)

    original_name, original_ext = src.stem, src.suffix
    dest_path = dest / src.name

    ######
    file_name_from_dir = list(src.parts[-3:-1])
    file_name_from_original_name = Path(src.name).with_suffix('.json').name # Change file suffix

    new_file_name = '_'.join(file_name_from_dir + [file_name_from_original_name])
    dest_path = dest_path.with_name(new_file_name)

    return dest_path
```

```python
files = get_all_files_in_dir_recursively(src, ['xml'])
```

```python
copy_all_files_in_one_folder(files, dest, custom_name_ext_process)
```
